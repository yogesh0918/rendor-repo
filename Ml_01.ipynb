```python
# Diabetes Prediction - Supervised ML

#### Import required libraries

import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

print("pandas=="+pd.__version__)
print("numpy=="+np.__version__)
print("scikit-learn=="+ sklearn.__version__)

#### import data - PIMA diabetes Dataset

db_data = pd.read_csv("C:/Users/dell/Desktop/D/Data-Analyst/ML/Projects/Diabetes-prediction/rendor-repo/Raw-data/diabetes.csv")
db_df = pd.DataFrame(db_data)
db_df.head()
```

```python

db_df.shape

db_df.describe()

db_df.info()

db_df.isnull().sum()

db_df['Outcome'].value_counts()

db_df.groupby('Outcome').mean()

#### Split Data - Train & Test

X = db_df.drop(columns = "Outcome", axis = 1)
Y = db_df["Outcome"]

X.shape, Y.shape

"""
Using Fit transofrm - Becuase value in all the features/coulumn have different ranges of data,
so we're standardizing the data and we've to work on standardize data (if values in same range - outcome will be more accurate)
so we're fit and transforming that data.
"""

sd_scalar = StandardScaler()

sd_scalar.fit(X)
X_transformed = sd_scalar.transform(X)

X_transformed

X_train, X_test, Y_train, Y_test = train_test_split(X_transformed, Y, test_size = 0.2, stratify=Y, random_state=2)

X_transformed.shape ,X_train.shape, X_test.shape, Y_train.shape, Y_test.shape

#### Model Training

classifier = svm.SVC(kernel='linear')

classifier.fit(X_train, Y_train)

#### Model evaluation

# Model accuracy
def accuracy_score_(Model,X, Y,train_test = ""):
  X_prediction = Model.predict(X)
  X_accuracy = accuracy_score(X_prediction, Y)
  print(f"Accuracy of {train_test} Data:", X_accuracy)

# Training Dataset accuracy
accuracy_score_(classifier,X_train, Y_train, "Training")

# Test Dataset accuracy
accuracy_score_(classifier,X_test, Y_test, "Test")

```
